```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read and load data from given linear data values
d_rd=pd.read_csv("/content/HW2_linear_data.csv")
x=d_rd.iloc[:,0]
y=d_rd.iloc[:,1]
plt.scatter(x,y,s=20)
plt.show()
```

```python
# Define linear model
def linear_model(X, m, c):
  return m * X + c
# Define mean square error(MSE) loss function
def mse_loss(Y_true, Y_pred):
  return np.mean((Y_true - Y_pred)**2)
m=0
c=0
L=0.0001
epochs=1000
n=len(x)
for i in range(epochs):
  # Calculate predictions
  y_pred=linear_model(x,m,c)
  # Calculate the gradient of the loss with respect to each parameter
  lf_m=(-2/n)*sum(x*(y-y_pred))
  lf_c=(-2/n)*sum(y-y_pred)
  # Update the parameters
  m=m-L*lf_m
  c=c-L*lf_c
  # Calculate the current loss
  curr_lost = mse_loss(y, y_pred)
  if i % 100 == 0:
    print(f"Epoch {i}: Loss = {curr_lost:.6f}")

print("Training completed.")
```

```python
# Plot the data and the fitted linear function using straight line equation
y_pred=m*x+c
plt.scatter(x,y,s=20)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)],color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.show()
```
